```r
#read dataset and pass into data frame
abalone <- read.csv("https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data", header=FALSE)

#add column names to data set
names(abalone) <- c("sex", "lenght", "diameter", "height", "weight.whole", "weight.shucked", "weight.viscera", "weight.shell", "rings")
```

```r
# number of rows in data set
nrow(abalone)
# show structure of dataset
str(abalone)
```

```r
# check for missing values
sum(is.na(abalone))
```

```r
# summary of dataset
summary(abalone)
```

```r
#ggplot to look at predictor variables and outcome variables
qplot(weight.shell, rings, data=abalone)
```

```r
# fit plot into OLS regression line
qplot(weight.shell, rings, data=abalone, geom=c("point", "smooth"), method="lm", color=sex, se=F)
```

```r
# this is a regression for each ses. shows the releationship between sex and shell weight
ggplot(abalone, aes(weight.shell, rings, color=sex)) + geom_jitter(alpha=0.5) + geom_smooth(method=lm, se=FALSE)
```

```r
# plotting each sex seprately
ggplot(abalone, aes(weight.shell, rings)) + geom_jitter(alpha=0.25) + geom_smooth(method=lm, se=FALSE) + facet_grid(.~ sex)
```
